```python
from langchain_core.prompts import PromptTemplate

# 1. Define the template string with a placeholder for the input variable
template_str = "Explain {topic} in simple terms for beginners"

# 2. Create the PromptTemplate object
# We explicitly define 'topic' as our input variable
task1_prompt = PromptTemplate(
    input_variables=["topic"],
    template=template_str
)

# 3. Test the template
# We use the .format() method to pass our dynamic variable instead of hardcoding
topic_input = "Artificial Intelligence"
formatted_prompt_1 = task1_prompt.format(topic=topic_input)

print("Task 1 Output:")
print(formatted_prompt_1)
```

```python
# 1. Define a template with three variables [cite: 59]
multi_input_template = "Explain {topic} for {audience} in a {tone} tone"

# 2. Create the PromptTemplate
task2_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=multi_input_template
)

# 3. Define the test cases provided in the assignment [cite: 60]
test_cases = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"}, # [cite: 62]
    {"topic": "Python", "audience": "kids", "tone": "fun"},       # [cite: 63]
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"} # [cite: 64]
]

print("Task 2 Outputs:")
# 4. Loop through the test cases and format the prompt for each
for test in test_cases:
    # **test unpacks the dictionary directly into the format parameters
    formatted_prompt = task2_prompt.format(**test)
    print(f"- {formatted_prompt}")
```

```python
# 1. Define the three different templates [cite: 67, 68, 69]
templates = {
    "Teaching": "Explain {topic} clearly step by step",
    "Interview": "Ask 3 questions about {topic}",
    "Storytelling": "Explain {topic} as a story" # [cite: 71]
}

# 2. Convert them into LangChain PromptTemplate objects
prompt_variations = {}
for style, temp_str in templates.items():
    prompt_variations[style] = PromptTemplate(
        input_variables=["topic"],
        template=temp_str
    )

# 3. Set the required topic [cite: 72]
current_topic = "Machine Learning"

print("Task 3 Outputs:")
# 4. Run the engine to generate all variations for the topic
for style, prompt_obj in prompt_variations.items():
    result = prompt_obj.format(topic=current_topic)
    print(f"[{style}] -> {result}")
```

```python
from langchain_core.prompts import ChatPromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate

# 1. Define the templates for the different roles
# The System message defines the behavior (role) [cite: 75, 76]
system_template = "You are a highly skilled {role}."
system_message_prompt = SystemMessagePromptTemplate.from_template(system_template)

# The Human message handles the dynamic input (topic) [cite: 77, 78]
human_template = "Please tell me about {topic}."
human_message_prompt = HumanMessagePromptTemplate.from_template(human_template)

# 2. Combine them into a single ChatPromptTemplate
chat_prompt = ChatPromptTemplate.from_messages([
    system_message_prompt,
    human_message_prompt
])

# 3. Test with the example provided [cite: 83, 84, 85]
role_input = "teacher"
topic_input = "Neural Networks"

# Format the final chat messages
formatted_chat = chat_prompt.format_messages(role=role_input, topic=topic_input)

print("Task 4 Output:")
for msg in formatted_chat:
    print(f"{msg.type.capitalize()}: {msg.content}")
```

```python
# 1. Define the validation function [cite: 89, 90]
def validate_inputs(audience, tone):
    # Allowed values based on rules [cite: 92]
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    # 2. Check if the inputs are valid
    if audience.lower() not in valid_audiences:
        # Raise an error if invalid [cite: 93, 94]
        raise ValueError(f"Invalid audience: '{audience}'. Must be one of {valid_audiences}")

    if tone.lower() not in valid_tones:
        raise ValueError(f"Invalid tone: '{tone}'. Must be one of {valid_tones}")

    # If everything is correct, return the validated inputs
    return audience.lower(), tone.lower()

# 3. Let's test the validation
print("Task 5 Test:")
try:
    # This should pass
    valid_a, valid_t = validate_inputs("beginner", "fun")
    print(f"Success! Validated: {valid_a}, {valid_t}")

    # This should trigger our error handling [cite: 94]
    validate_inputs("expert", "angry")
except ValueError as e:
    print(f"Error caught: {e}")
```

```python
# 1. Define the main generation function [cite: 99]
def generate_prompt(topic, audience, tone, style):
    try:
        # Step A: Validate the inputs using our Task 5 function
        val_audience, val_tone = validate_inputs(audience, tone)

        # Step B: Define our styles using PromptTemplate [cite: 101, 102]
        style_templates = {
            "teaching": PromptTemplate(
                input_variables=["topic", "audience", "tone"],
                template="Explain {topic} for {audience} in a {tone} teaching style"
            ),
            "interview": PromptTemplate(
                input_variables=["topic", "audience", "tone"],
                template="Ask an interview question about {topic} suitable for a {audience} candidate using a {tone} tone"
            ),
            "storytelling": PromptTemplate(
                input_variables=["topic", "audience", "tone"],
                template="Explain {topic} for {audience} in a {tone} storytelling style" # [cite: 104]
            )
        }

        # Step C: Ensure the requested style exists
        if style.lower() not in style_templates:
            raise ValueError(f"Invalid style. Choose from {list(style_templates.keys())}")

        # Step D: Format and return the final prompt
        selected_template = style_templates[style.lower()]
        return selected_template.format(
            topic=topic,
            audience=val_audience,
            tone=val_tone
        )

    except ValueError as e:
        return f"App Error: {e}"

# 2. Test the App with the example output requirement [cite: 103, 104]
print("Task 6 Output:")
app_result = generate_prompt(
    topic="Neural Networks",
    audience="beginner",
    tone="fun",
    style="storytelling"
)
print(app_result)
```
